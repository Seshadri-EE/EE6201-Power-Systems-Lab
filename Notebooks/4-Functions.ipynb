#### Scope of variables

The scope of a variable refers to the region of the code where the variable is accessible. In Python, there are two main types of variable scopes:

**Local Scope**: Variables defined within a function are said to have a local scope. They can only be accessed within that function.

**Global Scope**: Variables defined outside of any function have a global scope. They can be accessed from anywhere in the code, including inside functions.
